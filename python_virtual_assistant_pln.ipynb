#M√≥dulo 1 ‚Äì Text to Speech (tts.py)

Respons√°vel por converter texto em √°udio.

```python
from gtts import gTTS
from IPython.display import Audio, display
import os
import uuid

def falar(texto):
    nome_arquivo = f"fala_{uuid.uuid4()}.mp3"
    tts = gTTS(text=texto, lang="pt-br")
    tts.save(nome_arquivo)
    display(Audio(nome_arquivo, autoplay=True))
```

#M√≥dulo 3 ‚Äì Processamento de Comandos (comandos.py)

Interpreta o texto e executa a√ß√µes automatizadas.

```python
import webbrowser
import wikipedia
from tts import falar

wikipedia.set_lang("pt")

def executar_comando(comando):
    if "wikipedia" in comando:
        falar("Pesquisando na Wikipedia")
        termo = comando.replace("wikipedia", "").strip()

        try:
            resultado = wikipedia.summary(termo, sentences=2)
            falar(resultado)
        except:
            falar("N√£o encontrei resultados para essa pesquisa.")

    elif "youtube" in comando:
        falar("Abrindo o YouTube")
        webbrowser.open("https://www.youtube.com")

    elif "farm√°cia" in comando:
        falar("Mostrando a farm√°cia mais pr√≥xima")
        webbrowser.open("https://www.google.com/maps/search/farm√°cia+pr√≥xima")

    elif "sair" in comando:
        falar("Encerrando o assistente. At√© mais!")
        return False

    else:
        falar("Comando n√£o reconhecido.")

    return True
```

#M√≥dulo Principal (main.py)

Integra todos os m√≥dulos e mant√©m o assistente em execu√ß√£o.

```python
from tts import falar
from stt import ouvir_comando
from comandos import executar_comando

def main():
    falar("Ol√°! Eu sou seu assistente virtual. Como posso ajudar?")

    ativo = True
    while ativo:
        comando = ouvir_comando()
        if comando:
            ativo = executar_comando(comando)

if __name__ == "__main__":
    main()
```
